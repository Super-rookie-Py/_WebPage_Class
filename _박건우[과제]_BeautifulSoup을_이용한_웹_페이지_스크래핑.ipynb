## 네이버 도서 정보 스크래핑

- https://book.naver.com/ 네이버 책

```
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
from urllib.parse import quote

def get_books(query):

  book_df = pd.DataFrame(columns = ("Title", "Author", "Translator", "Publisher", "Lowest Price", "Book Introduce"))
  url_query = quote(query)
  url = 'https://book.naver.com/search/search.nhn?sm=sta_hty.book&sug=&where=nexearch&query=' + url_query
  idx = 0
  
  search_url = urllib.request.urlopen(url).read()
  soup = BeautifulSoup(search_url, 'html.parser')
  link_set = soup.select('#searchBiblioList > li > dl > dt > a')

  for link in link_set:
    try:
      title_link = link.get('href')
      search_title = urllib.request.urlopen(title_link).read()
      title_soup = BeautifulSoup(search_title, "html.parser")
      title = title_soup.select('#container > div.spot > div.book_info > h2 > a')[0].get_text()
      publisher = title_soup.find('a', {'class': 'N=a:bil.publisher'}).get_text()
      book_info = title_soup.select('#bookIntroContent > p')[0].get_text()
      lowest_price = title_soup.find('div',{'class':'lowest'}).find('strong').get_text()
      info = title_soup.select('#container > div.spot > div.book_info > div.book_info_inner > div')
      author = info[1].find('a').get_text()
      translator = info[1].find_all('a')[1].get_text()
    except:
      continue  
    book_df.loc[idx] = [title, author, translator, publisher, lowest_price, book_info]
    idx += 1
    print("#",end='')
  
  return book_df
```

```
query = input("검색 질의: ")
book_df = get_books(query)
print("Done")
```
